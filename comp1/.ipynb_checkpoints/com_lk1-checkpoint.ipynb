```python
import os
import pandas as pd

from matplotlib_venn import venn2
from matplotlib_venn import venn3
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
os.listdir('../proc2_S7A')
```

```python
# Criteria
var_tp_ml = 35 # The plot isn't centered on 50%
var_tp_mh = 55
var_tp_s = 30
var_t_rl = 400
var_t_rh = 5000
var_tv_nl = 100
var_tv_nh = 400
var_cg_pl = 4.0 # pipeline min is 3
var_cg_ph = 20
```

```python
cmap1 = mpl.colors.LinearSegmentedColormap.from_list("", ["green","dimgrey","purple"])
```

```python
# S7A dataframe
s7a_df = pd.read_csv('../proc2_S7A/cw-cvg_mer.csv')
s7a_df['t_r'] = s7a_df['tm_t'] + s7a_df['tu_t']
#
s_s7a_df = s7a_df
s_s7a_df = s_s7a_df[s_s7a_df['tp_m'] > var_tp_ml]
s_s7a_df = s_s7a_df[s_s7a_df['tp_m'] < var_tp_mh]
s_s7a_df = s_s7a_df[s_s7a_df['tp_s'] < var_tp_s]
s_s7a_df = s_s7a_df[s_s7a_df['tv_n'] > var_tv_nl]
s_s7a_df = s_s7a_df[s_s7a_df['tv_n'] < var_tv_nh]
s_s7a_df = s_s7a_df[s_s7a_df['t_r'] > var_t_rl]
s_s7a_df = s_s7a_df[s_s7a_df['t_r'] < var_t_rh]
s_s7a_df = s_s7a_df[s_s7a_df['cg_pct'] > var_cg_pl]
s_s7a_df = s_s7a_df[s_s7a_df['cg_pct'] < var_cg_ph]
s_s7a_df = s_s7a_df.dropna(subset=['c_nam'])
#
gl_s7a = str(s_s7a_df['c_nam'].unique())
gl_s7a = gl_s7a.replace("[", "")
gl_s7a = gl_s7a.replace("]", "")
gl_s7a = gl_s7a.replace("\"", "")
gl_s7a = gl_s7a.replace("\n", "")
gl_s7a = gl_s7a.replace(",", "")
gl_s7a = gl_s7a.replace("'", "")
gl_s7a = gl_s7a.split(' ')
gl_s7a = list(set(gl_s7a))
gl_s7a = sorted(gl_s7a)
print(len(gl_s7a))
print(gl_s7a)
```

```python
plt.scatter(s7a_df['tp_m'], s7a_df['tp_s'], s=1, alpha=0.25, c=s7a_df['tp_m'], cmap=cmap1)
plt.scatter(s_s7a_df['tp_m'], s_s7a_df['tp_s'], c='black', s=10, alpha=0.6)
plt.title("Echidna (S7A) CpG islands")
plt.xlabel("Average methylation (%)")
plt.ylabel("Variation of methylation (%)")
plt.vlines(var_tp_ml, 0, var_tp_s, 'k')
plt.vlines(var_tp_mh, 0, var_tp_s, 'k')
plt.hlines(var_tp_s, var_tp_ml, var_tp_mh, 'k')
plt.hlines(0, var_tp_ml, var_tp_mh, 'k')
plt.savefig('s7a_df_CGI-meth.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.scatter(s7a_df['tv_n'], s7a_df['t_r'], s=1, alpha=0.25)
plt.title("Echidna (S7A) CGI coverage")
plt.xlabel("CpG sites")
plt.ylabel("Meth + Unmeth counts")
plt.yscale('log')
plt.vlines(var_tv_nl, var_t_rl, var_t_rh, 'k')
plt.vlines(var_tv_nh, var_t_rl, var_t_rh, 'k')
plt.hlines(var_t_rl, var_tv_nl, var_tv_nh, 'k')
plt.hlines(var_t_rh, var_tv_nl, var_tv_nh, 'k')
plt.savefig('s7a_df_CGI-reads.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.scatter(s7a_df['cg_pct'], s7a_df['tp_m'], s=1, alpha=0.25)
plt.title("Echidna (S7A) CpG percent")
plt.xlabel("CpG percent")
plt.ylabel("Average methylation (%)")
plt.hlines(var_tp_ml, var_cg_pl, var_cg_ph, 'k')
plt.hlines(var_tp_mh, var_cg_pl, var_cg_ph, 'k')
plt.vlines(var_cg_pl, var_tp_ml, var_tp_mh, 'k')
plt.vlines(var_cg_ph, var_tp_ml, var_tp_mh, 'k')
plt.savefig('s7a_df_CGI-pct.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
cnam = 'OSBPL9' #'YPEL4'
fig.suptitle(cnam, y=1.05)
ss_s7a_df = s7a_df[s7a_df['c_nam'].notna()]
ss_s7a_df = ss_s7a_df[ss_s7a_df['c_nam'].str.contains(cnam)]
sss_s7a_df = s_s7a_df[s_s7a_df['c_nam'].str.contains(cnam)]
ss_t8a_df = t8a_df[t8a_df['c_nam'].notna()]
ss_t8a_df = ss_t8a_df[ss_t8a_df['c_nam'].str.contains(cnam)]
sss_t8a_df = s_t8a_df[s_t8a_df['c_nam'].str.contains(cnam)]
ss_pla_df = pla_df[pla_df['c_nam'].notna()]
ss_pla_df = ss_pla_df[ss_pla_df['c_nam'].str.contains(cnam)]
sss_pla_df = s_pla_df[s_pla_df['c_nam'].str.contains(cnam)]
ax1.set_title('Echidna (S7A)')
ax1.scatter(s7a_df['tp_m'], s7a_df['tp_s'], s=1, alpha=0.25, c=s7a_df['tp_m'], cmap=cmap1)
ax1.scatter(ss_s7a_df['tp_m'], ss_s7a_df['tp_s'], c='blue', s=10, alpha=0.6)
ax1.scatter(sss_s7a_df['tp_m'], sss_s7a_df['tp_s'], c='black', s=20, alpha=0.6)
ax2.set_title('Echidna (T8A)')
ax2.scatter(t8a_df['tp_m'], t8a_df['tp_s'], s=1, alpha=0.25, c=t8a_df['tp_m'], cmap=cmap1)
ax2.scatter(ss_t8a_df['tp_m'], ss_t8a_df['tp_s'], c='blue', s=10, alpha=0.6)
ax2.scatter(sss_t8a_df['tp_m'], sss_t8a_df['tp_s'], c='black', s=20, alpha=0.6)
ax3.set_title('Platypus')
ax3.scatter(pla_df['tp_m'], pla_df['tp_s'], s=1, alpha=0.25, c=pla_df['tp_m'], cmap=cmap1)
ax3.scatter(ss_pla_df['tp_m'], ss_pla_df['tp_s'], c='blue', s=10, alpha=0.6)
ax3.scatter(sss_pla_df['tp_m'], sss_pla_df['tp_s'], c='black', s=20, alpha=0.6)
plt.savefig('cpgmeth_'+cnam+'.png', dpi=300, bbox_inches='tight')
plt.show()
```
